## File: AutoParts Sample Data

### CASE 1
### Variables: Revenue VS Employee, Country Code, IndustryID, isActive

```python
# Importing the dataset  
dataset = pd.read_csv(r'C:\Users\DT\Desktop\자료\Sample Data Autoparts.csv')  
```

```python
x = dataset.iloc[:, [7, 8, 20, 21,22]].values
print(x)
```

```python
#finding optimal number of clusters using the elbow method  
from sklearn.cluster import KMeans  
wcss_list= []  #Initializing the list for the values of WCSS  
  
#Using for loop for iterations from 1 to 10.  
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x)  
    wcss_list.append(kmeans.inertia_)  
mtp.plot(range(1, 11), wcss_list)  
mtp.title('The Elobw Method Graph')  
mtp.xlabel('Number of clusters(k)')  
mtp.ylabel('wcss_list')  
mtp.show()  
```

```python
#training the K-means model on a dataset  
kmeans = KMeans(n_clusters=3, init='k-means++', random_state= 42)  
y_predict= kmeans.fit_predict(x)  
```

```python
#visulaizing the clusters  
mtp.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 50, c = 'red', label = 'Cluster 1') #for first cluster  
mtp.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 50, c = 'green', label = 'Cluster 2') #for second cluster 
mtp.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], s = 50, c = 'cyan', label = 'Cluster 3') #for fourth cluster
mtp.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 50, c = 'black', label = 'Centroid')   
mtp.title('Case 1')  
mtp.xlabel('employee_cnt (1 person)')  
mtp.ylabel('revenue')  
mtp.legend()  
mtp.show()
```
